```python
import requests
from urllib.parse import quote

def openalex_work_by_title(title: str, api_key: str):
    q = quote(f'"{title}"')
    url = f"https://api.openalex.org/works?search={q}&per-page={1}"
    r = requests.get(url, params={"api_key": api_key}, timeout=30)
    r.raise_for_status()
    return r.json()["results"]
```
